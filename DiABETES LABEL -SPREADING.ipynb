```python
X=merged[['DIABETES','CHOLESTEROL_DISORDERS','CHRONIC_KIDNEY_DISEASE','FATTY_LIVER','HYPERTRIGLYCERIDEMIA','OBESITY']]
y=merged[['Diabetes']]
```

```python
y
```

```python
import numpy as np

y_set1 = y.copy()
y_set1['Diabetes'] = np.select(
    [
        y_set1['Diabetes'].isin(['Low','Low to Mild']),
        y_set1['Diabetes'].isin(['Mild']),
        y_set1['Diabetes'].isin(['Mild to Moderate','Moderate']),
        y_set1['Diabetes'].isin(['Moderate to High'])
    ],
    ['Low', 'Mild','Moderate','Moderate to high'],
    default= y_set1['Diabetes']
)


y_set1['Diabetes'].replace('0', -1, inplace=True)
y_set1['Diabetes'].replace('-',-1,inplace=True)
```

```python
y_set1['Diabetes'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode labels for categorical values
encoder = LabelEncoder()
y_encoded_set1 = y_set1['Diabetes'].copy()

# Replace -1 (unlabeled) with NaN for encoding
y_encoded_set1.replace(-1, None, inplace=True)

# Apply LabelEncoder to known labels
encoded_labels = encoder.fit_transform(y_encoded_set1.dropna())

# Assign encoded labels back to the original DataFrame using the index
y_encoded_set1[y_encoded_set1.notna()] = encoded_labels

# Convert back to -1 for unlabeled
y_encoded_set1.fillna(-1, inplace=True)
print("Encoded Labels: ", y_encoded_set1.unique())
```

```python
from sklearn.semi_supervised import LabelSpreading

# Prepare labels (converting to integers for LabelSpreading)
labels_combined = y_encoded_set1.values.astype(int)

# Create the Label Spreading model
label_spread = LabelSpreading(kernel='knn', alpha=0.2)  # You can adjust 'alpha' and 'kernel' as needed

# Fit the model to the features (X) and the labels (labels_combined)
label_spread.fit(X.values, labels_combined)

# Get the propagated labels
y_encoded_set1_spread = label_spread.transduction_

# Calculate accuracy on known labels (non-unlabeled)
known_indices = y_encoded_set1 != -1
accuracy_spreading = (y_encoded_set1[known_indices] == y_encoded_set1_spread[known_indices]).mean()
print("Label Spreading Accuracy on Known Labels:", accuracy_spreading)
```

```python
# Step 1: Identify Unlabeled Data
unlabeled_indices = (y_encoded_set1 == -1)

# Step 2: Extract Propagated Labels for Unlabeled Data
propagated_labels_for_unlabeled_spread = y_encoded_set1_spread[unlabeled_indices]

# Step 3: Display Propagated Labels for Unlabeled Data
print("Propagated Labels for Unlabeled Data (Label Spreading):")
print(propagated_labels_for_unlabeled_spread)
```

```python
import pandas as pd

# Step 1: Copy the original DataFrame
df_with_spread_labels = y_set1.copy()  # Assuming y_set4 is the original data

# Step 2: Add Propagated Labels (from Label Spreading) to the DataFrame
df_with_spread_labels['Propagated Labels'] = y_encoded_set1_spread.copy()  # Use the propagated labels from Label Spreading

# Step 3: Update the unlabeled data with the propagated labels
df_with_spread_labels.loc[unlabeled_indices, 'Propagated Labels'] = propagated_labels_for_unlabeled_spread  # Propagated labels for unlabeled data

# Step 4: Convert the 'Propagated Labels' to integer type
df_with_spread_labels['Propagated Labels'] = df_with_spread_labels['Propagated Labels'].astype(int)

# Step 5: Print a sample of the updated DataFrame
print(df_with_spread_labels.head())

# Step 6: Save the updated DataFrame to a new CSV file
# df_with_spread_labels.to_csv('/content/drive/MyDrive/MUSCULOSKELETAL HEALTH/updated_with_propagated_labels_spread1111.csv', index=False)
```
